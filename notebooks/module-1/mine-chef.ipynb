```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model('deepseek-v4-flash' , extra_body = {
    "thinking" : {
        "type" : "disabled"
    }
})
```

```python
from langchain.tools import tool
from tavily import TavilyClient
from typing import Dict , Any

tavily_client = TavilyClient()

@tool
def search_tool(query : str) -> Dict[str , Any]:
    """ searches web for the information"""
    return tavily_client.search(query)
```

```python
system_prompt = """

You are a personal chef. The user will give you a list of ingredients they have left over in their house.

Using the web search tool, search the web for recipes that can be made with the ingredients they have.

Return recipe suggestions and eventually the recipe instructions to the user, if requested.

"""
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model= model,
    tools=[search_tool],
    system_prompt= system_prompt,
    checkpointer= InMemorySaver()
)

 
```

```python
from langchain.messages import HumanMessage

question = HumanMessage("i have chicken breats 400g some cabbage , beans , capsicum , ketchup , soya sauce and some other sauces and masalas left in kitchen tell me what should i make ")
config = {"configurable" : {"thread_id" : "1"}}

response = agent.invoke(
    {"messages" : [question]},
    config
)
```

```python
from pprint import pprint
pprint(response["messages"][-1].content)
```

```python
from langchain.messages import HumanMessage
from pprint import pprint

response = agent.invoke(
    {"messages" : HumanMessage(content="what all sauces apart from soya sauce do i need")},
    config
)

pprint(response["messages"][-1].content)
```
