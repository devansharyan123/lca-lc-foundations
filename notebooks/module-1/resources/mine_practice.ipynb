```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model('deepseek-v4-flash' , extra_body = {
    "thinking" : {
        "type" : "disabled"
    }
})
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage , SystemMessage
from pydantic import BaseModel , Field
from pprint import pprint

class Beaches(BaseModel):
    name_of_beach : str
    city  : str
    state : str
    country : str

class BeachesResponse(BaseModel):
    beaches: list[Beaches] = Field(min_length=3, max_length=3)

question1 = SystemMessage(content="You are a travel guide. Return exactly 3 of the most beautiful beaches in the country the user asks for.")    

agent = create_agent(model = model , 
                     system_prompt=question1
                     ,response_format=BeachesResponse
                     )   

question = HumanMessage(content="What are the most beautiful beaches of the Italy")

response = agent.invoke(
    {"messages" : [question]}
)
pprint(response["structured_response"])
```

```python
index = response["structured_response"]

for beach in index.beaches:
    print(beach.name_of_beach + " : " + beach.city + " , " + beach.country)
```
